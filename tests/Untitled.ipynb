```python
def block_average(block_size, values, weights):
    n = len(values)
    xb = values.reshape((n // block_size, block_size))
    wb = weights.reshape((n // block_size, block_size))
    means = [np.average(x, weights=w) for x, w in zip(xb, wb)]
    weights = np.sum(wb, axis=1)
    print(np.average(means, weights=weights), np.std(means), len(means))
```

```python
for i in [10, 16, 20, 25, 40, 50, 80, 100, 125]:
    block_average(i, x, w)
```
